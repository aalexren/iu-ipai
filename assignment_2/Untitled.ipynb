```python
from mido import Message, MidiTrack, MetaMessage, MidiFile
import math

def make_chord(note_on: Message, note_off: Message, length: int) -> list[Message]:
    # make major chord Б3 + М3
    
    return [
    Message('note_on', channel = 0, note = note_on.note - 24, velocity = 50, time = 0),
    Message('note_on', channel = 0, note = note_on.note - 24 + 4, velocity = 50, time = 0),
    Message('note_on', channel = 0, note = note_on.note - 24 + 7, velocity = 50, time = 0),
    
    Message('note_off', channel = 0, note = note_off.note - 24, velocity = 50, time = length),
    Message('note_off', channel = 0, note = note_off.note - 24 + 4, velocity = 50, time = 0),
    Message('note_off', channel = 0, note = note_off.note - 24 + 7, velocity = 50, time = 0)
    ]

def get_midi_time(file: MidiFile) -> int:
    return sum([note.time for track in mid.tracks for note in track if type(note) is Message and note.type in ['note_on', 'note_off']])

def save_midi(tracklist: list[MidiTrack], name='new_song.mid'):
    midi = mido.MidiFile(type = 1)
    for track in tracklist:
        midi.tracks.append(track)
    midi.save(name)
    return midi

melody: list[Message] = [note for track in mid.tracks for note in track if type(note) is Message and note.type in ['note_on', 'note_off']]
track_mel = MidiTrack()
track_acc = MidiTrack()
track_meta = MidiTrack()

track_meta.append(MetaMessage('time_signature', numerator=4, denominator=4, clocks_per_click=24, notated_32nd_notes_per_beat=8, time=0))
track_meta.append(MetaMessage('set_tempo', tempo=480000, time=0))

chord_count = math.ceil(get_midi_time(midi) / midi.ticks_per_beat) 
chord_time = midi.ticks_per_beat

for i in range(len(melody) - 1):
    track_mel.append(melody[i])

    chord = make_chord(melody[i], melody[i+1], chord_time)
    for c in chord:
        track_acc.append(c)

print(save_midi([track_meta, track_mel, track_acc]))
```
